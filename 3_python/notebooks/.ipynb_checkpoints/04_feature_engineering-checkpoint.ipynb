# Prepare data for Power BI & future ML

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\Prafull Wahatule\Desktop\Customer-Churn-Analysis\1_data\processed\customer_churn_cleaned.csv")
```

### STEP 3.4.1 – CREATE FEATURES

```python
df['high_support'] = (df['support_calls'] >= 5).astype(int)
df['late_payment'] = (df['payment_delay'] >= 10).astype(int)
df['low_usage'] = (df['usage_frequency'] <= 5).astype(int)
```

### STEP 3.4.2 – SAVE FEATURE DATASET

```python
df.to_csv(r"C:\Users\Prafull Wahatule\Desktop\Customer-Churn-Analysis\1_data\processed\customer_churn_features.csv", index=False)

print("✅ Prepare data for Power BI & future ML completed")
```
